# Benchmark results for removing duplicated items from a collection

```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [10, 5]
```

```python
data = json.load(open('../results/benchmark_unique_o2.json'))
benchmarks_raw = pd.DataFrame(data['benchmarks'])
benchmarks_raw.rename({'per_family_instance_index': 'data_size', 'family_index': 'benchmark_name'}, axis=1, inplace=True)
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(0, '1 KB')
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(1, '10 KB')
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(2, '100 KB')
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(3, '1 MB')
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(4, '10 MB')
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(5, '100 MB')
benchmarks_raw['data_size'] = benchmarks_raw['data_size'].replace(6, '1 GB')

benchmarks_raw['benchmark_name'] = benchmarks_raw['benchmark_name'].replace(0, 'STD_Sort_Unique')
benchmarks_raw['benchmark_name'] = benchmarks_raw['benchmark_name'].replace(1, 'HashSet_Remove_Dup')
benchmarks_raw['benchmark_name'] = benchmarks_raw['benchmark_name'].replace(2, 'TreeSet_Remove_Dup')
benchmarks_raw['benchmark_name'] = benchmarks_raw['benchmark_name'].replace(3, 'Vector_Remove_Dup')
benchmarks_raw['benchmark_name'] = benchmarks_raw['benchmark_name'].replace(4, 'List_Remove_Dup')

#benchmarks_raw['cpu_time'] = benchmarks_raw['cpu_time'].replace(2260925.977, np.nan)
#benchmarks_raw['cpu_time'] = benchmarks_raw.apply(lambda x: "{:,}".format(x['cpu_time']), axis=1)

benchmarks_raw
```

## CPU time (ms)

```python
benchmarks = benchmarks_raw.pivot_table('cpu_time', ['data_size', ], 'benchmark_name')
benchmarks = benchmarks.reindex(['1 KB', '10 KB', '100 KB', '1 MB', '10 MB', '100 MB', '1 GB'], axis=0)
pd.options.display.float_format = "{:,.3f}".format
benchmarks
```

```python
ax = benchmarks.plot()
plt.show()
```
